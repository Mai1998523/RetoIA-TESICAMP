```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import random as rd
from geopy import distance
from random import sample

data = pd.read_csv('coordenadas.csv') 
```

```python
def comparacion(poblacion1):
    l = tuple(poblacion1)
    comparacion = []
    for i in range(len(poblacion1)):
        g = l[i][30-1][3] #REgistro,#fila#columna
        comparacion.append((('Ruta ',i),('Distancia',g))) 
        #comparacion.append('Ruta {} con total de distancia {}'.format(i,g))
        #df = pd.DataFrame(data=p, columns=['Colonia1','Coloni2','Distacia/Colonia','Sumadistancia'])
        #df
    return comparacion

def comparacion2(poblacion2):
    l = tuple(poblacion2)
    comparacion2 = []
    for i in range(len(poblacion2)):
        g2 = l[i][29-1][3] #REgistro,#fila#columna
        comparacion2.append((('Ruta ',i),('Distancia',g2))) 
        #comparacion.append('Ruta {} con total de distancia {}'.format(i,g))
        #df = pd.DataFrame(data=p, columns=['Colonia1','Coloni2','Distacia/Colonia','Sumadistancia'])
        #df
    return comparacion2

#1
poblacion1 = []
poblacion2 = []
nPoblacion = 5

#2
while (len(poblacion1)<nPoblacion):
    data=data.sample(len(data)).reset_index(drop=True)
    grupo1 = np.array(data[['Colonia','Latitud','Longitud']][0:31])
    grupo1.shape
    m = 0
    pob1 = []
    if grupo1[0,0]=='TESI'and grupo1[30,0]=='TESI':
        for i in range(len(grupo1)-1): #fila
            coloniaA = (grupo1[i,1], grupo1[i,2])
            coloniaB = (grupo1[i+1,1], grupo1[i+1,2])
            distancia = distance.distance(coloniaA,coloniaB).km
            m = m + distancia
            pob1.append(((grupo1[i,0]),(grupo1[i+1,0]), distancia,m)) 
            #print(i,"La distancia del {} a {} es de {}km ---> DistanciaT:{} km".format((grupo1[i,0]),(grupo1[i+1,0]),distancia,m))
        poblacion1.append(pob1)
while (len(poblacion2)<nPoblacion):
    data=data.sample(len(data)).reset_index(drop=True)
    grupo2 = np.array(data[['Colonia','Latitud','Longitud']][32:])
    grupo2.shape
    mm = 0
    pob2 = []
    if grupo2[0,0]=='TESI'and grupo2[29,0]=='TESI' :
        for i in range(len(grupo2)-1): #fila
            coloniaAA = (grupo2[i,1], grupo2[i,2])
            coloniaBB = (grupo2[i+1,1], grupo2[i+1,2])
            distanciaa = distance.distance(coloniaAA,coloniaBB).km
            mm = mm + distanciaa
            pob2.append(((grupo2[i,0]),(grupo2[i+1,0]), distanciaa,mm)) 
            #print(i,"La distancia del {} a {} es de {}km ---> DistanciaT:{} km".format((grupo2[i,0]),(grupo2[i+1,0]),distancia,m))
        poblacion2.append(pob2)
print("AUTOBUS 1",comparacion(poblacion1))
print("AUTOBUS 2",comparacion2(poblacion2))
```
